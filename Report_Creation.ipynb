# Code to Create Report for One Table
* Replace 'INSERT_TABLE_NAME_HERE' with correct table name & replace 'INSERT_SCHEMA_NAME_HERE' with correct schema name

```python
# CODE FOR ONE TABLE 

# Create connection to Snowflake using snowflake.connector.connect() and authenticate using externalbrowser
conn = snowflake.connector.connect(
        account = account,
        user = user,
        authenticator = authenticator,
        warehouse = warehouse,
        database = database,
        role = role,
        schema = 'INSERT_SCHEMA_NAME_HERE')

# Create SQL query to run
sql = 'SELECT TOP 10000 * FROM INSERT_TABLE_NAME_HERE'

# Execute SQL query
cursor = conn.cursor()
cursor.execute(sql)

try: 
    df = cursor.fetch_pandas_all()
    cursor.close()
    report= create_report(df)
    report.save('INSERT_TABLE_NAME_HERE.html')
except:
    print('Failed to read table/view')

# Place HTML files into designated folder
source_folder = os.getcwd()
destination_folder = 'GVR_DEV/INSERT_SCHEMA_NAME_HERE'

for file_name in os.listdir(source_folder):
    if file_name.endswith('.html'):
        shutil.move(os.path.join(source_folder, file_name), destination_folder)

# Close the connection
conn.close()
```
